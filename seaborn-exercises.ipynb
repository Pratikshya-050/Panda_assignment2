```python
Ques 3: dataset-- titanic 
    Create a bar chart for the number of male and female passengers.
        ## output should be as shown below-
```

```python
Ques 4: dataset-- titanic
    Set up a grid to plot survival probability against several variables like class, sex, who and alone.
    Then Draw a seaborn pointplot onto each Axes
        ## output should be as shown below-
```

```python
sns.set_style('darkgrid')

fig,axes = plt.subplots(1,4,figsize=(12,4))
sns.pointplot(x='class',y='survived',data=titanic,ax=axes[0],color='maroon')

sns.pointplot(x='sex',y='survived',data=titanic,ax=axes[1],color='maroon')

sns.pointplot(x='who',y='survived',data=titanic,ax=axes[2],color='maroon')

sns.pointplot(x='alone',y='survived',data=titanic,ax=axes[3],color='maroon')

plt.tight_layout()
```

```python
Ques 5: dataset-- diamonds
    Generate the distribution of prices with respect to color for each cut quality.
    ## output should be as shown below
```

```python
diamonds=sns.load_dataset('diamonds')
sns.set_style('white')
sns.barplot(x='cut', y='price', data=diamonds,hue='color')
plt.legend(loc='upper right',ncol=4)
```

```python
Ques 6: dataset-- diamonds
    obtain the counts of diamonds of each cut quality through bar chart.
    ## output should be as shown below
```

```python
sns.countplot(x='cut',data=diamonds,color='red')
plt.xlabel('cut')
plt.ylabel('count')
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
Ques 7: dataset-- diamonds
Create a histogram of the frequency of diamonds in the dataset with their respective carat specifications on the x-axis.
## output should be as shown below
```

```python
sns.histplot(diamonds['carat'],bins=30,kde=False)
sns.set_style("white")
```

```python
Ques 8: dataset-- tips
    Draw a nested violinplot and split the violins for easier comparison between day and total_bill.
    ## output should be as shown below
```

```python
tip=sns.load_dataset("tips")
sns.violinplot(x='day', y='total_bill', data=tip,split=True,hue='sex',inner='quartile')
sns.set_palette('Set3')
sns.set_style("whitegrid")
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Violin Plot: Total Bill by Day')
plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
Ques 10: dataset-- fmri()
    Create line plot to show the responses for different events and regions
    ## output should be as shown below
```

```python
fmri=sns.load_dataset('fmri')
sns.relplot(x="timepoint", y="signal", hue="region", style="event",kind="line", data=fmri)
sns.set_style("darkgrid")
```

```python
#Ques 11: dataset-- iris
#    Create a scatter plot between sepal length and sepal width categorised by species.
            ## output should be as shown below
iris = sns.load_dataset('iris')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
sns.set_style("white")
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
Ques 15: dataset-- exercise
    Draw a pointplot to show pulse as a function of three categorical factors rest, walking and running.
    ## output should be as shown below
```

```python
exercise = sns.load_dataset("exercise")
filtered_data = exercise.loc[exercise['diet'].isin(['low fat', 'no fat'])]
sns.catplot(x='time', y='pulse', hue='kind', col='diet', kind='point', data=filtered_data,color='blue')
```
